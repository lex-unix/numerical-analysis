```python
import os
import numpy as np
import pandas as pd
```

```python
A = np.array([
    [2.67, 1.71, 5.42],
    [-6.39, 1.45, 3.84],
    [4.21, 5.92, -3.41]
])

x = np.ones(len(A)).T
e = 10e-6
```

```python
i = 0
l = 0
error = 0
Lambdas = []
Vectors = np.zeros(3)
Errors = []
while True:
    y = A.dot(x)
    c = max(y, key=abs)
    y = (1 / c) * y
    x = y
    error = abs(l - c)
    l = c
    Lambdas.append(l)
    Vectors = np.vstack((Vectors, x))
    Errors.append(error)
    if error < e:
        break
    i += 1
```

```python
print(l, x)
```

```python
data = {
    'lambda': Lambdas,
    'u1': Vectors[1:, 0],
    'u2': Vectors[1:, 1],
    'u3': Vectors[1:, 2],
    'e': Errors
}
```

```python
df = pd.DataFrame(data)
```

```python
df.index = np.arange(1, len(df) + 1)
pd.set_option('display.float_format', lambda x: '%.7f' % x)
```

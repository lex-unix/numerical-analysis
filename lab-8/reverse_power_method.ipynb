```python
import os
import numpy as np
import pandas as pd
```

```python
A = np.array([
    [4.30, 4.62, 4.10, 1.90],
    [3.92, 8.45, 5.78, 2.46],
    [3.77, 9.21, 8.04, 2.28],
    [2.21, 1.65, 1.69, 1.19]
])

x = np.ones(len(A)).T
alpha = 0.5
e = 10e-6
```

```python
i = 0
l = 0
I = np.identity(len(A)) * alpha
A = A - I
A = np.linalg.inv(A)
error = 0
Lambdas = []
Vectors = np.array(x)
Errors = []
while True:
    y = np.dot(A, x)
    c = np.max(np.absolute(y))
    y = (1 / c) * y
    x = y
    error = abs(l - c)
    l = c
    Lambdas.append(l)
    Vectors = np.vstack((Vectors, x))
    Errors.append(error)
    if error < e:
        break
    i += 1
```

```python
l = alpha + 1 / l
Lambdas[-1] = l
```

```python
data = {
    'lambda': Lambdas,
    'u1': Vectors[1:, 0],
    'u2': Vectors[1:, 1],
    'u3': Vectors[1:, 2],
    'u4': Vectors[1:, 3],
    'errors': Errors
}
```

```python
df = pd.DataFrame(data)
```

```python
df.index = np.arange(1, len(df) + 1)
pd.set_option('display.float_format', lambda x: '%.7f' % x)
```
